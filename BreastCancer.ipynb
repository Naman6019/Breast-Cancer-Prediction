# Perform upsampling or down-sampling to the dataset in order to train the model properly

```python
data_frame.groupby('label').mean()
```

```python
X = data_frame.drop(columns='label',axis=1)
Y = data_frame['label']
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.3, random_state=2)
```

```python
print(X.shape, X_test.shape, X_train.shape)
```

```python
#Perform scalar standardization on data to amplify accuracy
from sklearn.preprocessing import StandardScaler
```

Dont fit test data only training data

```python
scalar = StandardScaler()

X_train_std = scalar.fit_transform(X_train)
X_test_std = scalar.transform(X_test)
```

# Building Neural Network
Importing TensorFlow and Keras

Keras is included in tesorflow library. These both are used to create nueral networks. We have used randon.set_seed so that, every run of the model will give us the same accuracy score. Or else, we would get different accuracy score for the model for every iteration.

```python
import tensorflow as tf
tf.random.set_seed(3)
from tensorflow import keras
```

1. Flatten coverts the numerical data into 1D array from an matrix to feed into the input layer. Here we have 30 input nuerons
2. Dense is the hidden layer taking the input from all the input layer nodes with an activation function. Here we use relu(Rectified Linear Units)
3. Dense(the 2nd one) is the output layer with 2 nuerons and the activation function sigmoid(in this case).
4. The number of output layer nuerons depend upon the number of different classes in the dataset.

```python
#setting up the layers of the nueral network
model = keras.Sequential([
                        keras.layers.Flatten(input_shape=(30,)), 
                        keras.layers.Dense(20, activation='relu'),
                        keras.layers.Dense(2, activation='sigmoid')])
```

```python
#Compiling the nueral network
model.compile(optimizer='adam', loss='sparse_categorical_crossentropy', metrics=['accuracy'])
```

validation_split = same as test_train_split function
epochs = number of iterations

```python
#training the nueral network
history = model.fit(X_train_std, Y_train, validation_split=0.1, epochs=20)
```

```python
plt.plot(history.history['accuracy'])
plt.plot(history.history['val_accuracy'])

plt.title('model accuracy')
plt.ylabel('accuracy')
plt.xlabel('epoch')

plt.legend(['training data', 'validation data'], loc = 'lower right')
```

```python
plt.plot(history.history['loss'])
plt.plot(history.history['val_loss'])

plt.title('model loss')
plt.ylabel('loss')
plt.xlabel('epoch')

plt.legend(['training data', 'validation data'], loc = 'upper right')
```

```python
loss, accuracy = model.evaluate(X_test_std, Y_test)
print(loss,'________',accuracy)
```

```python
Y_pred = model.predict(X_test_std)
```

The output for the following cell will give the probablity for the datapoint being of class 0 or 1, whereas in ML, we will get output as 1 or 0 everytime.

np.argmax function will iterate over the whole array to get the index for the max value in the list

```python
#Converting the prediction probavlity to labels
Y_pred_labels = [np.argmax(i) for i in Y_pred]
print(Y_pred_labels)
```
